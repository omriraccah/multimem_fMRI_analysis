```python
import numpy as np
import matplotlib.pyplot as plt
import nilearn as ni
from nilearn import plotting, image
import nibabel as nib
```

```python
# directory
bids_dir = "/Users/aryanagarwal/Downloads/sample_data_GLM/"
sub = "mm01"
condition = "visual-audio"

# Paths
t1w_path = f"{bids_dir}preprocessed/sub-{sub}/anat/sub-{sub}_desc-preproc_T1w.nii.gz"  
mni_path = f"{bids_dir}sandbox/atlas/mni_icbm152_t1_tal_nlin_sym_09a.nii.gz"
z_map_t1w_path = f"{bids_dir}sandbox/GLM_structs/{condition}/T1w/contrast-{condition}_sub-{sub}_runs-all_space-T1w.nii.gz"   
z_map_mni_path = f"{bids_dir}sandbox/GLM_structs/{condition}/MNI152NLin2009cAsym/contrast-{condition}_sub-{sub}_runs-all_space-MNI152NLin2009cAsym.nii.gz"

# Load images
z_map_t1w = image.load_img(z_map_t1w_path)
z_map_mni = image.load_img(z_map_mni_path)
t1w_img = image.load_img(t1w_path)
mni_img = image.load_img(mni_path)

# Plot
plotting.plot_stat_map(z_map_mni, bg_img=mni_img, threshold=3, display_mode='mosaic', title=f"{condition}")
plotting.show()
```

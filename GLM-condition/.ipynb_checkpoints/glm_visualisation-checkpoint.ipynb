```python
def plot_roi_correlations(df, x_col='roi', y_col='corr', hue_col='cond', 
                          roi_order=None, condition_order=None, 
                          palette=None, ylim=None, yticks=None,
                          custom_legend_labels=None, title=None,
                          fig_height=6, fig_aspect=2.5,
                          ylabel="Mean Model Accuracy ($\mathbf{z}$)"):
    """
    Create a beautiful bar plot of correlation differences across ROIs with different conditions.
    
    Parameters:
    -----------
    df : pandas DataFrame
        DataFrame containing the data to plot
    roi: str
        The roi to plot. Will filter df accordingly. Use short name.
    x_col : str, default='kernel'
        Column name for x-axis categories (ROIs)
    y_col : str, default='acc'
        Column name for y-axis values (correlation differences)
    hue_col : str, default='cond'
        Column name for grouping variable (conditions)
    roi_order : list, optional
        Specific order for ROIs on x-axis
        Default: ["HPC", "CA1", "CA2+3", "DG", "EC", "PRC", "Subiculum"]
    condition_order : list, optional
        Specific order for conditions in the legend
        Default: ["A-A", "A-IC_A", "V-V", "V-IC_V", "C-C", 'V-A']
    palette : str or list, optional
        Color palette for the plot (defaults to custom scientific palette)
    ylim : tuple, optional
        Y-axis limits as (min, max)
    yticks : list, optional
        Custom y-tick positions
    custom_legend_labels : list, optional
        Custom labels for the legend
    title : str, optional
        Title for the plot
    fig_height : int, default=6
        Figure height
    fig_aspect : float, default=2.5
        Figure aspect ratio (width/height)
    ylabel : str, default="Mean Pattern Similarity ($\mathbf{z}$)"
        Label for the y-axis
    
    Returns:
    --------
    matplotlib.axes.Axes
        The plot axes
    """
    # Filter the dataframe
    df = df[df['sub'] != 'mm04']
    df = df[df['sub'] != 'mm12']
    
    if condition_order is None:
        condition_order = ["A","V", "C"]

    # Set default palette if not provided
    if palette is None:
        # Modern colorblind-friendly palette without brown
        palette = ['#4e79a7', '#f28e2b', '#59a14f', '#e15759', '#76b7b2', '#edc949']
        # Alternative options:
        # Blue-focused: palette = ['#003f5c', '#2f4b7c', '#665191', '#a05195', '#d45087', '#f95d6a']
        # Tableau-inspired: palette = ['#4e79a7', '#f28e2b', '#59a14f', '#e15759', '#76b7b2', '#edc949']
    
    # Set the style for a clean, professional look
    plt.rcParams.update({
        'font.family': 'sans-serif',
        'font.sans-serif': ['Arial', 'Helvetica', 'DejaVu Sans'],
        'font.size': 12,
        'axes.titlesize': 16,
        'axes.labelsize': 14,
        'xtick.labelsize': 12,
        'ytick.labelsize': 12,
        'legend.fontsize': 11,
        'figure.figsize': (fig_height * fig_aspect, fig_height)
    })
    
    # Use a clean style with minimal gridlines
    sns.set_style("ticks")
    
    # Create figure for more control
    fig = plt.figure(figsize=(fig_height * fig_aspect, fig_height), dpi=100)
    
    # Create the categorical plot with improved design
    ax = sns.barplot(
        x=x_col, 
        y=y_col, 
        hue=hue_col, 
        data=df, 
        order=roi_order,
        hue_order=condition_order,  # This ensures colors are assigned in the correct order
        palette=palette,
        alpha=0.85,  # Slight transparency for more aesthetic look
        errorbar=('ci', 95),  # Updated from deprecated 'ci' parameter
        errwidth=1.2,
        capsize=0.05,
        errcolor="black",
        ax=plt.gca()
    )
    
    # Customize the grid for a cleaner look
    ax.yaxis.grid(True, linestyle='--', alpha=0.7, color='#cccccc')
    ax.set_axisbelow(True)  # Put grid behind bars
    
    # Remove top and right spines for cleaner look
    sns.despine()
    
    # Set pure white background
    ax.set_facecolor('white')
    fig.patch.set_facecolor('white')
    
    # Enhance bar edges
    for patch in ax.patches:
        patch.set_edgecolor('black')
        patch.set_linewidth(0.8)
    
    # Set axis labels with proper formatting
    ax.set_xlabel("", fontweight='bold')
    ax.set_ylabel(ylabel, fontweight='bold')  # Use the customizable y-axis label
    
    # Set y-axis limits if provided
    if ylim:
        ax.set_ylim(ylim)
    
    # Set custom y-ticks if provided
    if yticks:
        ax.set_yticks(yticks)
    
    # Add title if provided
    if title:
        ax.set_title(title, fontweight='bold', pad=15)
    
    # Remove the default legend
    if ax.get_legend() is not None:
        ax.get_legend().remove()
    
    # Add a custom, better positioned legend
    if custom_legend_labels:
        handles, _ = ax.get_legend_handles_labels()
        # Create a legend with properly matched colors
        legend = ax.legend(
            handles, 
            custom_legend_labels,
            frameon=True,
            framealpha=0.95,
            edgecolor='#cccccc',
            loc='upper right',
            bbox_to_anchor=(1.01, 1.01)
        )
    else:
        # Create default legend
        legend = ax.legend(
            frameon=True,
            framealpha=0.95,
            edgecolor='#cccccc',
            loc='upper right', 
            bbox_to_anchor=(1.01, 1.01)
        )
    legend.get_frame().set_linewidth(0.8)
    
    # Improve x-tick appearance
    plt.setp(ax.get_xticklabels(), rotation=0, ha='center', fontweight='bold')
    
    # Add a subtle box around the plot
    for spine in ax.spines.values():
        spine.set_linewidth(0.8)
    
    # Adjust layout
    plt.tight_layout()
    
    return ax
```

```python
# Helper to strip the paths of the stimuli to create numbered conditions for GLMSingle
def strip_paths(list_o_paths):
    stripped = []
    for s in list_o_paths:
        if 'Incongruent' not in s:
            stripped.append([int(s.split('/S')[-1].split('_')[0])])
        else:
            temp = [int(s.split('/S')[-1].split('_')[i].replace('S', '')) for i in [0, 1]]
            stripped.append(temp)
    return stripped
```

```python
# Function to make the design matrices
def make_design_matrices(func_dir, sub, nruns, conds):
    """
    :param func_dir: String path to preprocessed data
    :param sub: String subject number "mm01"
    :param nruns: Int number of runs
    :param conds: List of conditions: "Audio", "Visual", "Congruent", "Incongruent"
    :return: List of nruns design matrices
    """

    # initialize list for storing matrices across runs
    designmats = []
    
    # exlusion df
    exclusion_df = pd.read_csv('/gpfs/milgram/scratch60/turk-browne/or62/sandbox/decoding_structs/greater_than_1_5_exclusion.csv')
    exclusion_df_sub = exclusion_df[exclusion_df['subs'] == f"mm{sub}"]

    # Loop over runs and process events files, turn into design matrices, append
    for run in range(1, nruns + 1):
        # load events data for current run
        events_path = f"{func_dir}/sub-mm{sub}_task-multisensorymemory_run-{run}_events.tsv"
        events = pd.read_csv(events_path, sep='\t')

        # load confounds file for current run
        confounds, _ = ni.interfaces.fmriprep.load_confounds(f"{func_dir}/sub-mm{sub}_task-multisensorymemory_run-{run}_space-T1w_desc-preproc_bold.nii.gz", strategy=['high_pass', 'motion', 'scrub'], motion='basic', scrub=0, fd_threshold=0.5, std_dvars_threshold=1.5)
        
        # load exlusion trials for current run
        exclusion_df_sub_run = exclusion_df_sub[exclusion_df_sub['run'] == run]
        
        if exclusion_df_sub_run['stim_past_thr'].empty == False:
            trials_to_exclude = ast.literal_eval(exclusion_df_sub_run['stim_past_thr'].iloc[0])
        else:
            trials_to_exclude = []

        # adapt trial type labels to reflect conditions
        for index, row in events.iterrows():
            if '/Audio/' in events['trial_type'][index]:
                
                num = strip_paths([events['trial_type'][index][2:-2]])
                num = num[0][0]
                trial_name = f"S{num}_A"
                print
                if trial_name in trials_to_exclude:
                    events.loc[index, 'trial_type'] = "Excluded"
                else:
                    events.loc[index, 'trial_type'] = "Audio"
                    
            elif '/Visual/' in events['trial_type'][index]:
                num = strip_paths([events['trial_type'][index][2:-2]])
                num = num[0][0]
                trial_name = f"S{num}_V"
                
                if trial_name in trials_to_exclude:
                    events.loc[index, 'trial_type'] = "Excluded"
                else:
                    events.loc[index, 'trial_type'] = "Visual"
                    
            elif '/Congruent/' in events['trial_type'][index]:
                num = strip_paths([events['trial_type'][index][2:-2]])
                num = num[0][0]
                trial_name = f"S{num}_C"
                
                if trial_name in trials_to_exclude:
                    events.loc[index, 'trial_type'] = "Excluded"
                else:
                    events.loc[index, 'trial_type'] = "Congruent"
                    
            elif '/Incongruent/' in events['trial_type'][index]:
                num = strip_paths([events['trial_type'][index][2:-2]])
                num = num[0]
                trial_name = f"S{num[0]}_S{num[1]}_IC"
                
                if trial_name in trials_to_exclude:
                    events.loc[index, 'trial_type'] = "Excluded"
                else:
                    events.loc[index, 'trial_type'] = "Incongruent"

        # define conditions for GLM analysis
        events = events[events['trial_type'].isin(conds)]

        # change all durations to six seconds to account for slight discrepancies in task readout
        events['duration'] = events['duration'].replace(events['duration'].unique(), 6)

        # Set Frame Times
        t_r = 1.5
        n_scans = 337 if sub == "01" else 335
        frame_times = (np.arange(n_scans) * t_r)

        # create first-level design matrix
        designmat = make_first_level_design_matrix(
            frame_times,
            events,
            drift_model=None,  # loading it from fmriprep instead
            # drift_order=3,
            add_regs=confounds,
            # high_pass=high_pass,
            hrf_model="glover + derivative + dispersion")

        designmats.append(designmat)

    return designmats
```

```python
preproc_dir = f"/gpfs/milgram/scratch60/turk-browne/or62/sandbox/preprocessed"
sub = "04"
nruns = 9
conds = ["Audio", "Visual", "Congruent", "Incongruent", "Excluded"]
design_matrices = make_design_matrices(f"{preproc_dir}/sub-mm{sub}/func/", sub, nruns, conds)
```

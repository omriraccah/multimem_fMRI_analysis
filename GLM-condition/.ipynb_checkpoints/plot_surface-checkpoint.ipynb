```python
def plot_vol_to_surf(result_vol, cmap_type='default', output_file=None, threshold=None, vmax=None, title=None, cmap='magma', cbar_range=None, cbar_cmap=None):
    assert cmap_type in ['divergent','mirror','default']

    if cmap_type == 'divergent':
        cmap = get_merged_VWRDPU_cmap()
    elif cmap_type == 'mirror':
        cmap = mirror_cmap(plt.get_cmap(cmap))
    else:
        cmap = plt.get_cmap(cmap)

    from matplotlib.colors import ListedColormap
    import matplotlib.gridspec as gridspec

    # First, create the brain surface plot normally (this will create its own figure)
    plotting.plot_img_on_surf(result_vol, 
                              views=['lateral','medial'], 
                              cmap=cmap,
                              inflate=True, 
                              bg_on_data=False, 
                              alpha=0.8,
                              darkness=0.8, 
                              vmax=vmax,
                              title=title,
                              threshold=threshold,
                              colorbar=False
                             )

    # Now get the current figure and modify it
    fig = plt.gcf()

    # Adjust the existing brain plots to make room for colorbar
    # This shrinks the brain plots vertically to 85% of original size
    for ax in fig.axes:
        pos = ax.get_position()
        ax.set_position([pos.x0, pos.y0 + 0.1, pos.width, pos.height * 0.85])

    # Create a new axes for the horizontal colorbar at the bottom
    cbar_ax = fig.add_axes([0.15, 0.02, 0.7, 0.05])  # [left, bottom, width, height]

    # Use specified colorbar colormap or default to original cmap
    if cbar_cmap is not None:
        original_cmap = plt.get_cmap(cbar_cmap)
    else:
        original_cmap = plt.get_cmap(cmap)  # Original, non-mirrored

    # TRUNCATE the colormap to show only the significant portion
    cmap_start = 0.25  # Start at 25% of the colormap (skip the dark bottom)
    cmap_end = 1.0     # Use up to 100% of the colormap

    # Create a truncated colormap
    import numpy as np
    n_colors = 256
    colors = original_cmap(np.linspace(cmap_start, cmap_end, n_colors))
    truncated_cmap = ListedColormap(colors)

    # Use the truncated colormap
    norm = Normalize(vmin=threshold, vmax=vmax)
    mappable = cm.ScalarMappable(norm=norm, cmap=truncated_cmap)
    mappable.set_array([])

    # Create HORIZONTAL colorbar in the dedicated space
    cbar = plt.colorbar(mappable, cax=cbar_ax, orientation='horizontal')

    # Convert threshold and vmax back to p-values
    p_min = 10**(-vmax)      # smallest p-value (0.0001)
    p_max = 10**(-threshold) # largest p-value (0.05)

    # Create tick positions
    p_tick_values = np.logspace(np.log10(p_min), np.log10(p_max), 6)
    tick_positions = -np.log10(p_tick_values)

    cbar.set_ticks(tick_positions)

    # Format p-value labels
    p_labels = []
    for p_val in p_tick_values:
        if p_val >= 0.01:
            p_labels.append(f'{p_val:.2f}')
        elif p_val >= 0.001:
            p_labels.append(f'{p_val:.3f}')
        elif p_val >= 0.0001:
            p_labels.append(f'{p_val:.4f}')
        else:
            p_labels.append(f'{p_val:.0e}')

    cbar.set_ticklabels(p_labels)

    # Beautiful styling for the HORIZONTAL colorbar
    cbar.ax.tick_params(
        labelsize=11,
        length=5,
        width=1.2,
        colors='#333333',
        direction='out',
        pad=6
    )

    # Elegant colorbar label (horizontal placement)
    cbar.set_label(
        'p-value', 
        labelpad=15,
        fontsize=12,
        fontweight='600',
        color='#333333'
    )

    # Beautiful border styling
    cbar.outline.set_linewidth(1.2)
    cbar.outline.set_edgecolor('#666666')

    # Style the tick labels for consistency
    for label in cbar.ax.get_xticklabels():
        label.set_fontweight('500')
        label.set_color('#333333')
        label.set_fontsize(11)

    # Optional: Add subtle grid for better readability
    cbar.ax.grid(True, which='major', alpha=0.2, linestyle='-', linewidth=0.5, color='gray')
    
def mirror_cmap(cmap):
        
    return merge_cmaps(cmap, cmap, 'mirrored_cmap',mirror1=True)

def merge_cmaps(first_cmap, second_cmap, merged_cmap_name, mirror1=False):
    '''
    this function assumes you're passing two objects of matplotlib.colors.LinearSegmentedColormap

    '''
    colors1 = first_cmap(np.linspace(0, 1, 128))
    if mirror1: colors1=colors1[::-1]
    colors2 = second_cmap(np.linspace(0, 1, 128))
    colors_combined = np.vstack((colors1,colors2))
    cmap_combined = ListedColormap(colors_combined, name=merged_cmap_name)
    return cmap_combined

def modify_colorbar_range(cbar_min, cbar_max):
    """Modify the colorbar to show only a specific range"""
    fig = plt.gcf()
    for ax in fig.get_axes():
        try:
            ax.set_ylim(cbar_min, cbar_max)
            # If we get here without error and the axis has some content, it might work
            if len(ax.get_children()) > 0:
                return
        except:
            continue
```

```python
# File paths and parameters
glm_dir = "/gpfs/milgram/scratch60/turk-browne/or62/sandbox/GLM_structs"
contrast = "incongruent-congruent"
fn = f"{glm_dir}/{contrast}/MNI152NLin2009cAsym/randomise_output_{contrast}_MNI152NLin2009cAsym_tfce_corrp_tstat1.nii.gz"
# Define views with more comprehensive coverage
views = ['lateral', 'medial']
# Load and transform the data
nii = nib.load(fn)

nii = image.math_img('-np.log10(1-X)', X=nii)
# Define p-value thresholds and their -log10 equivalents
p_threshold = 0.05
p_max = 0.001
log_threshold = -np.log10(p_threshold)
log_max = -np.log10(p_max)


plot_vol_to_surf(nii, cmap_type="mirror", threshold=log_threshold, output_file=None, title=contrast, vmax=log_max, cmap='magma',cbar_range=True, cbar_cmap='magma')
```

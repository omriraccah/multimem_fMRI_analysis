```python
# Load Dependencies
import warnings
import sys 
if not sys.warnoptions:
    warnings.filterwarnings("ignore", category=DeprecationWarning)
import os 
import nibabel as nib
import numpy as np
import nilearn as ni
import scipy.io
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.preprocessing import StandardScaler
import pandas as pd
from nilearn.plotting import plot_surf_stat_map
from nilearn import plotting
import glmsingle
from glmsingle.glmsingle import GLM_single
from matplotlib import pyplot as plt
from pprint import pprint
from nilearn.masking import apply_mask
from nilearn.plotting import plot_roi
from nilearn.image import index_img, resample_to_img
from nilearn.maskers import NiftiSpheresMasker, NiftiMasker
from os.path import join, exists, split
from itertools import count
import itertools

%matplotlib inline 
sns.set(style='white', context='notebook', rc={"lines.linewidth": 2.5})
sns.set(palette="colorblind")
```

```python
# Load GLM Single results for the final model
def load_glm_single(outputdir_glmsingle):
    results_glmsingle = dict()
    results_glmsingle['typed'] = np.load(join(outputdir_glmsingle,'TYPED_FITHRF_GLMDENOISE_RR.npy'),allow_pickle=True).item()
    
    betas = results_glmsingle['typed']['betasmd']
    
    return betas
```

```python
# function to strip the paths of the stimuli to create numbered conditions for GLMSingle
def strip_paths(list_o_paths):
    stripped = []
    for s in list_o_paths:
        if 'Incongruent' not in s:
            stripped.append([int(s.split('/S')[-1].split('_')[0])])
        else:
            temp = [int(s.split('/S')[-1].split('_')[i].replace('S', '')) for i in [0, 1]]
            stripped.append(temp)
    return stripped
```

```python
# Function to create condlist
def get_labels(nruns, preproc_dir, sub):

    # initialize list for storing matrices across runs
    trial_labels = pd.DataFrame()
    trial_labels_cond = pd.DataFrame()

    # initialize list for storing run labels
    run_labels = list()

    for run in range(1,nruns+1):
    
        # load events data for current run
        events_path = os.path.join(preproc_dir, f"sub-mm{sub}/func/sub-mm{sub}_task-multisensorymemory_run-{run}_events.tsv")
        events = pd.read_csv(events_path, sep='\t')

        # change all durations to six seconds to account for slight discrepecies in task readout
        events['duration'] = events['duration'].replace(events['duration'].unique(), 1.5)
    
        # remove any ITI events from matrix (slightly different for mm01 due to label issue)
        if sub == "01":
            events = events[~events['trial_type'].str.contains('MCOS')]
            events = events.reset_index(drop=True)
        else:
            events = events[~events['trial_type'].str.contains('ITI')]
            events = events.reset_index(drop=True)
    
        events["trial_type_label"] = events["trial_type"]
    
        for index, row in events.iterrows():
    
            if '/Audio/' in events['trial_type_label'][index]:

                num = strip_paths([events['trial_type_label'][index][2:-2]])
                num = num[0][0]
                events.loc[index, 'trial_type'] = f"S{num}_A"
                events.loc[index, 'trial_type_cond'] = "A"

            elif '/Visual/' in events['trial_type_label'][index]:

                num = strip_paths([events['trial_type_label'][index][2:-2]])
                num = num[0][0]
                events.loc[index, 'trial_type'] = f"S{num}_V"
                events.loc[index, 'trial_type_cond'] = "V"

            elif '/Congruent/' in events['trial_type_label'][index]:

                num = strip_paths([events['trial_type_label'][index][2:-2]])
                num = num[0][0]
                events.loc[index, 'trial_type'] = f"S{num}_C"
                events.loc[index, 'trial_type_cond'] = "C"

            elif '/Incongruent/' in events['trial_type_label'][index]:
                num = strip_paths([events['trial_type_label'][index][2:-2]])
                num = num[0]
                events.loc[index, 'trial_type'] = f"S{num[0]}_S{num[1]}_IC"
                events.loc[index, 'trial_type_cond'] = "IC"
        
            run_labels.append(run)
    
        # save run labels
        trial_labels[run] = events['trial_type']
        trial_labels_cond[run] = events['trial_type_cond']
    
    # stack all trials to one column (vertically) for each correspondence to GLM trial betas
    trial_labels = pd.concat([trial_labels, trial_labels.T.stack().reset_index(name='all_runs')['all_runs']], axis=1)
    trial_labels_cond = pd.concat([trial_labels_cond, trial_labels_cond.T.stack().reset_index(name='all_runs')['all_runs']], axis=1)
    condlist = pd.DataFrame()
    condlist['conds_spec'] = trial_labels['all_runs']
    condlist['conds_gen'] = trial_labels_cond['all_runs']
    condlist['runs'] = run_labels
    
    return condlist
```

```python
# Function to apply mask to betas
def extract_betas(sub, preproc_dir, roi, betas, hpc, pos):
    
    if hpc:
        roi_mask = f"{preproc_dir}/sub-mm{sub}/rois/ASHS/final/func_masks/bin_masks_5B_T1/{pos}_{roi}_mask_T1_non-bin.nii.gz"
    else:
        roi_mask = f"{preproc_dir}/sub-mm{sub}/rois/harvard-oxford/{roi}.nii.gz"

    func_path = f'{preproc_dir}/sub-mm{sub}/func/sub-mm{sub}_task-multisensorymemory_run-1_space-T1w_desc-preproc_bold.nii.gz'
    mask_nii = nib.load(roi_mask)
    func_nii = nib.load(func_path)

    # Run resampling here
    resampled_mask_nii = resample_to_img(mask_nii, func_nii, interpolation="nearest")

    # Get array, convert to boolean, mask
    mask_array = resampled_mask_nii.get_fdata()
    mask_bin = mask_array > 0
    masked_betas = betas[mask_bin].T
    
    return masked_betas
```

```python
# CODE TO CREATE MATRICES FOR EACH CONDITION (AVERAGED) FOR MASKED BETAS WE CAN THEN PERFORM VOXEL WISE NORMALIZATION
# Function to make the RSA matrix
def reconstruct_masked_data(conds, condlist, masked_betas, norm=True):

    # Panda to hold voxels (roi) x each condition scene (e.g., S1_V, S2_A, ...)
    pd.options.mode.chained_assignment = None
    masked_betas_restructured = pd.DataFrame()

    # loop through conditions
    for cond in conds:

        # loop through scenes
        for scene in range(1, 11):

            # get indeces for current scene in current condition
            ind = condlist['conds_spec'].str.contains(f'S{scene}_{cond}').to_list()

            # average betas across runs
            scene_avg = np.sum(masked_betas[ind], axis=0)/nruns

            masked_betas_restructured[f'S{scene}_{cond}'] = scene_avg

    if norm:
        condition_spec_normed = pd.DataFrame()

        # loop through to normalize
        for cond in conds:

            columns_with_substring = masked_betas_restructured.filter(like=f'_{cond}').columns
            condition_spec = masked_betas_restructured[columns_with_substring]

            # loop through rows of panda
            for index, row in condition_spec.iterrows():

                # normalize and return row
                scaler = StandardScaler()
                row_normed = scaler.fit_transform(row.values.reshape(-1, 1))
                condition_spec.iloc[index] = row_normed[:,0]

            # Concatenate horizontally
            condition_spec_normed = pd.concat([condition_spec_normed, condition_spec], axis=1)

        return condition_spec_normed

    else:
        return masked_betas_restructured
```

```python
# Function to make the RSA matrix
def get_rsa_matrix(conds, masked_betas_restructured):

    # Now we need to filter, sort, RSA, and visualise
    scene_id_1 = [f"S{i}_{cond}" for cond in conds[0] for i in range(1, 11)]
    scene_id_2 = [f"S{i}_{cond}" for cond in conds[1] for i in range(1, 11)]

    # Initialise RSA matrix in pandas
    rsa_corrs = pd.DataFrame(index=scene_id_1, columns=scene_id_2)

    for s in scene_id_1:
        for ss in scene_id_2:
            if s != ss:
            
                # Calculate correlation and store it
                s1_data = masked_betas_restructured[s].values
                s2_data = masked_betas_restructured[ss].values
                rsa_corrs.loc[s, ss] = np.corrcoef(s1_data, s2_data)[1, 0]

            else: # Diagonal
                rsa_corrs.loc[s, ss] = 0
    
    rsa_corrs = rsa_corrs[rsa_corrs.columns].astype(float)
    
    return rsa_corrs
```

```python
# Function to visualise RSA matrix
def visualise_rsa(rsa_corrs, roi):
    # Visualize the RSA matrix using seaborn
    plt.figure(figsize=(8, 8))
    sns.heatmap(rsa_corrs, center=0, cmap='coolwarm', annot=False, square=True, cbar=True,xticklabels=True, yticklabels=True)

    # Add title and labels
    plt.title(f"RSA Matrix Heatmap {roi}", fontsize=16)
    plt.xlabel("Stimulus ID", fontsize=12)
    plt.ylabel("Stimulus ID", fontsize=12)

    plt.tight_layout()
    plt.show()
    
    return 0
```

```python
# Function to run entire RSA pipeline
def run_rsa(preproc_dir, sub, roi, nruns, conds, hpc, vis, pos, norm):
    
    """
    preproc_dir: Path to the preprocessed directory (not subject specific!)
    sub: The subject to analyse as a string e.g. "04"
    roi: The name of the ROI to use, must be the FILENAME
    nruns: Number of runs to analyse (goes 1 to nruns)
    conds: Conditions to analyse, must be specified as list of strings
    hpc: Boolean that specifies if the ROI is from ASHS (True) or Harvard-Oxford (False - default)
    norm: Boolean specifies whether to perform voxel-wise normalization. 
    """
    
    # Get the GLM Single Betas
    betas = load_glm_single(f"{preproc_dir}/glm_single_results_mm{sub}/")
    
    # Get the associated labels
    condlist = get_labels(nruns, preproc_dir, sub)
    
    # Mask the betas
    masked_betas = extract_betas(sub, preproc_dir, roi, betas, hpc, pos)
    
    if conds[0] == conds[1]:
        masked_betas_normed = reconstruct_masked_data(conds[0], condlist, masked_betas, norm)
    else:
        masked_betas_normed = reconstruct_masked_data(conds, condlist, masked_betas, norm)
    
    # Get the RSA matrix
    rsa_corrs = get_rsa_matrix(conds, masked_betas_normed)
    
    # Visualise the matrix
    if vis: visualise_rsa(rsa_corrs, roi)
    
    return rsa_corrs
```

```python
# Set constants and variables required to run this analysis

# ROI specifications
rois = ["Heschl's Gyrus (includes H1 and H2)", "Occipital Pole", "Superior Temporal Gyrus, anterior division", "Superior Temporal Gyrus, posterior division", "HPC", "CA1", "CA2+3", "DG", "EC", "PHC", "PRC", "Subiculum"] # Has to be the name the file is saved as
rois_bool = [False, False, False, False, True, True, True, True, True, True, True, True] # Boolean to indicate if the program should look in ASHS
pos = ["", "", "", "", "combined", "combined", "combined", "combined", "combined", "combined", "combined", "combined"] # Orientation for ASHS outputs: combined/left/right

# The subject IDs to include
subs = ["01", "02", "03", "04", "06", "07"]

# Other constants
nruns = 9
norm = True
conds = ["A", "V", "C"]
user = "or62"
vis = False # Whether to plot rsa matrices 
preproc_dir = f"/gpfs/milgram/scratch60/turk-browne/{user}/sandbox/preprocessed/"

# Don't change, building interactions to loop over
conditions = ['AA', 'AV', 'VV', 'CC']

# Dataframe to store the final results
df = pd.DataFrame(columns=['sub', 'cond', 'roi', 'corr'])

# Loop over subjects
for sub in subs:
    
    # Loop over conditions
    for cond in conditions:
        
        # Loop over ROI
        for r in range(len(rois)):
            # Calculate the RSA matrix for given subject, condition, and ROI
            rsa_matrix = run_rsa(preproc_dir=preproc_dir, sub=sub, roi=rois[r], nruns=nruns, 
                                 conds=[cond[0], cond[1]], hpc=rois_bool[r], vis=vis, pos=pos[r], norm=norm)
            rsa_matrix = rsa_matrix.to_numpy()
            
            # Calculate the mean correlation for the interaction
            if cond[0] != cond[1]:
                mean = np.mean(rsa_matrix)
            # Exclude the diagonal when the interaction is between the same conditions
            else:
                mean = np.mean(rsa_matrix[~np.eye(len(rsa_matrix), dtype=bool)])
            
            # Store the mean in our dataframe
            df.loc[len(df)] = [sub, cond, rois[r], mean]
            print(f"Done with {rois[r]} for {cond} in {sub}")
        
            
df.head(n=10)        
```
